```python
import seaborn as sns
from matplotlib.pyplot import imshow

import numpy as np

from colormaps_dataset import cmap_dataframe, generate_cmap_probabilities
```

```python
red_proba, green_proba, blue_proba = generate_cmap_probabilities(exclude=["Qualitative", "Uncategorized"])
```

```python
red_proba.sum(axis=1)
```

```python
ax = sns.heatmap(red_proba, cmap="Reds")
ax.set_title("Red Probability Matrix")
ax.set_xlabel("Value Walked Too")
ax.set_ylabel("Value Walked From")
```

```python
ax = sns.heatmap(green_proba, cmap="Greens")
ax.set_title("Green Probability Matrix")
ax.set_xlabel("Value Walked Too")
ax.set_ylabel("Value Walked From")
```

```python
ax = sns.heatmap(blue_proba, cmap="Blues")
ax.set_title("Blue Probability Matrix")
ax.set_xlabel("Value Walked Too")
ax.set_ylabel("Value Walked From")
```

```python
df = cmap_dataframe()
int_df = cmap_dataframe(convert_data_to_int=True)

df[~df.category.isin(["Qualitative", "Uncategorized"])].cmap_name.unique()
```

```python
color_name = "viridis"
color_frame = df[df.cmap_name == color_name]
img1 = np.array([color_frame[["red", "green", "blue"]]])
ax = imshow(img1, aspect='auto')
```

```python
int_color_frame = int_df[int_df.cmap_name == color_name]
red, green, blue = int_color_frame.iloc[0][['red', 'green', 'blue']]
red, green, blue
```

```python
argmax_red = np.argmax(red_proba, axis=1)
argmax_green = np.argmax(green_proba, axis=1)
argmax_blue = np.argmax(blue_proba, axis=1)
test_colors = [(red/255, green/255, blue/255)]
for _ in range(256):
    red, green, blue = argmax_red[red], argmax_green[green], argmax_blue[blue]
    test_colors.append((red / 255, green / 255, blue / 255))
```

```python
ax = imshow(np.array(test_colors), aspect='auto')
```
